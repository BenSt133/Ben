```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy.random as random
import ipdb
import copy

#torch stuff
import torch
import pytorch_lightning as pl

#analysis / software for ml
from ana_lib.custom_plmodels import *
from ana_lib.optuna_utils import *
from ana_lib.noise import *
from ana_lib.pat_utils import *
from ana_lib.pl_callbacks import *
from ana_lib.pl_utils import *
from ana_lib.plot_utils import *
from ana_lib.process_spectrum import *
from ana_lib.torch_loader import *
from ana_lib.utils import *


import pandas
```

```python
data_load = torch.load('Test_fullsave_RunMarch23_test_EXP')
```

```python
xin = data_load['xin']
xout = data_load['xout']
xt = data_load['xt']
xskip = data_load['xskip']
xoutf = data_load['xoutf']
yt = data_load['yt']
xfinal = data_load['xfinal']
```

```python
mods = [0,1,2] #just plot these ones
mods_len = len(mods)+1

tnums = np.arange(0,4000,100)

for tnum in tnums:

    fig=plt.figure(figsize=[20,20])

    mod=15
    #Input MNIST image
    plt.subplot(mods_len,4,1)
    input_image = np.reshape(xt[tnum,:],(14,14))
    input_image = np.flipud(input_image)
    plt.pcolor(input_image)
    #plt.plot(xs[epoch][tnum,:],'.-',label='Input image')
    #plt.legend()

    #after manifold and concatenate
    plt.subplot(mods_len,4,2)
    #lt.plot(xin_dtsv[epoch][mod][tnum].detach().numpy(),'.-',label='Input train, Digital twin')
    plt.plot(xin[mod][tnum].detach().numpy(),'.-',label='Physical input')
    plt.legend()

    #output from physical system
    plt.subplot(mods_len,4,3)
    #lt.plot(xout_dtsv[epoch][mod][tnum].detach().numpy(),'.-',label='Output train, Digital twin')
    plt.plot(xout[mod][tnum].detach().numpy(),'.-',label='Physical output')    
    plt.legend()

    #after skip connection 
    plt.subplot(mods_len,4,4)
    plt.plot(xskip[mod][tnum].detach().numpy(),'.-',label='Skip')
    plt.plot(xout[mod][tnum].detach().numpy(),'.-',label='Fwd')    
    plt.legend()
    xpe = xskip[mod][tnum].detach().numpy() + xout[mod][tnum].detach().numpy()

    #next row
    mod=16
    #before manifold
    plt.subplot(mods_len,4,5)
    #lt.plot(xpd.detach().numpy(),'.-',label='Input train, Digital twin')
    plt.plot(xpe,'.-',label='Physical input') 

    #after manifold 
    plt.subplot(mods_len,4,6)
    #lt.plot(xin_dtsv[epoch][mod][tnum].detach().numpy(),'.-',label='Input train, Digital twin')
    plt.plot(xin[mod][tnum].detach().numpy(),'.-',label='Physical input') 

    #output from physical system
    plt.subplot(mods_len,4,7)
    #lt.plot(xout_dtsv[epoch][mod][tnum].detach().numpy(),'.-',label='Output train, Digital twin')
    plt.plot(xout[mod][tnum].detach().numpy(),'.-',label='Physical output')    
    plt.legend()

    #after skip connection 
    plt.subplot(mods_len,4,8)
    plt.plot(xskip[mod][tnum].detach().numpy(),'.-',label='Skip')
    plt.plot(xout[mod][tnum].detach().numpy(),'.-',label='Fwd')    
    plt.legend()
    xpe = xskip[mod][tnum].detach().numpy() + xout[mod][tnum].detach().numpy()     


    #next row
    mod=17
    #before manifold
    plt.subplot(mods_len,4,9)
    plt.plot(xpe,'.-',label='Physical input') 

    #after manifold 
    plt.subplot(mods_len,4,10)
    #lt.plot(xin_dtsv[epoch][mod][tnum].detach().numpy(),'.-',label='Input train, Digital twin')
    plt.plot(xin[mod][tnum].detach().numpy(),'.-',label='Physical input') 

    #output from physical system
    plt.subplot(mods_len,4,11)
    #lt.plot(xout_dtsv[epoch][mod][tnum].detach().numpy(),'.-',label='Output train, Digital twin')
    plt.plot(xout[mod][tnum].detach().numpy(),'.-',label='Physical output')    
    plt.legend()

    #after skip connection 
    plt.subplot(mods_len,4,12)
    plt.plot(xskip[mod][tnum].detach().numpy(),'.-',label='Skip')
    plt.plot(xout[mod][tnum].detach().numpy(),'.-',label='Fwd')    
    plt.legend()
    xpe = xskip[mod][tnum].detach().numpy() + xout[mod][tnum].detach().numpy()    


    #Look at final output
    mod=5
    plt.subplot(mods_len,4,13)
    #lt.plot(xoutd.detach().numpy(),'.-',label='Last, Digital twin')  
    plt.plot(xpe,'.-',label='Last')
    plt.legend()

    plt.subplot(mods_len,4,14)
    plt.plot(xoutf[mod][tnum].detach().numpy(),'.-',label='Last Summed')

    xe =xoutf[mod][tnum].detach().numpy()

    plt.plot(yt[tnum]*np.ones(2),np.array([np.amin(xe),np.amax(xe)]),'k:',label='Target')
    plt.legend()


    #average over all channels
    plt.subplot(mods_len,4,15)
    for nc in range(0,7):
        xe =xoutf[nc][tnum].detach().numpy()
        plt.plot(xe,'.-',label='Last Summed ' + str(nc))

    #total average
    plt.subplot(mods_len,4,16)
    plt.plot(xfinal[0][tnum],'.-',label='Last Summed')

    xe =xfinal[0][tnum].detach().numpy()

    plt.plot(yt[tnum]*np.ones(2),np.array([np.amin(xe),np.amax(xe)]),'k:',label='Target')
    plt.legend()    

    fig.patch.set_facecolor('white')
    plt.tight_layout()
    plt.savefig('figures/example full' + str(tnum) +'.png')
    plt.savefig('figures/example full' + str(tnum) +'.svg')
    plt.close()
```
